## Sin wave plotting

Now, ill be plotting a sin curve suing matplotlib and some support from numpy


````python
import matplotlib.pyplot as plt
import numpy as np

x_values = np.arange(0, 18, 0.0001)
y_values = np.sin(x_values)

plt.plot(x_values, y_values, marker='s', color='indigo') 
plt.title("Sin Curve")
plt.xlabel("X values")
plt.ylabel("Y values")
plt.grid(True)
plt.show()
````
